What is a meta class in python

In Python, a metaclass is the “class of a class”: it controls how classes themselves are created and behave, just like a class controls how instances are created. By default, all classes are created using the built‑in type metaclass, so when you write class MyClass: ..., Python actually calls type behind the scenes to construct MyClass.

You use a custom metaclass when you want to modify or enforce behavior at class‑creation time—for example, automatically adding methods, validating attributes, or changing how inheritance works—by defining a class that inherits from type and overriding __new__ or __init__, then passing it via the metaclass=... keyword in your class definition. Metaclasses are a powerful, but advanced, feature; they’re great for frameworks and DSL‑like code, but usually unnecessary for everyday application logic.
